```python
import re
from string import punctuation
def clean_tweets(tweet):
    # remove username
    while tweet.find('@') > -1:
        username_idx = tweet.find('@')
        if username_idx != -1:
            username_end_idx = tweet.find(':', username_idx)
            username = tweet[username_idx: username_end_idx]
            tweet = tweet.replace(username, '')
    tweet = tweet.replace('RT : ', '')
    # punctuations cleaning
    for p in punctuation:
        tweet = tweet.replace(p,'')
    tweet = tweet.replace('\n', '')
    # remove emojies
    
    return tweet

def remove_emojis(tweet):
    # Regular expression pattern to match emojis
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # Emojis 1
                               u"\U0001F300-\U0001F5FF"  # Emojis 2
                               u"\U0001F680-\U0001F6FF"  # Emojis 3
                               u"\U0001F1E0-\U0001F1FF"  # Flags
                               u"\U00002500-\U00002BEF"  # Chinese characters
                               u"\U00002702-\U000027B0"  # Dingbats
                               u"\U00002702-\U000027B0"  # Dingbats
                               u"\U000024C2-\U0001F251" 
                               u"\U0001f926-\U0001f937" 
                               u"\U00010000-\U0010ffff" 
                               u"\u2640-\u2642" 
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # Variety of Unicode symbols & text
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    
    # Remove emojis from the text
    tweet = emoji_pattern.sub(r'', tweet)
    
    return tweet
```

```python
import nltk
from nltk.corpus import opinion_lexicon
from nltk.corpus import words

def extract_english_words(text):
    # Load the English word corpus
    english_words = set(words.words())

    # Tokenize the input text into words
    tokens = nltk.word_tokenize(text)

    # Extract English words from the tokenized text
    english_word_list = [word for word in tokens if word.lower() in english_words]

    return english_word_list

def extract_negative_words(text):
    # Load the negative word lexicon
    negative_words = set(opinion_lexicon.negative())

    # Tokenize the input text into words
    tokens = nltk.word_tokenize(text)

    # Extract negative words from the tokenized text
    negative_word_list = [word for word in tokens if word.lower() in negative_words]

    return negative_word_list

def extract_positive_words(text):
    # Load the negative word lexicon
    negative_words = set(opinion_lexicon.positive())

    # Tokenize the input text into words
    tokens = nltk.word_tokenize(text)

    # Extract negative words from the tokenized text
    negative_word_list = [word for word in tokens if word.lower() in negative_words]

    return negative_word_list

# Example usage
negative_words = extract_positive_words(all_text)
print(negative_words)
```
