```python
import pandas as pd
import  numpy as np
import plotly.express as px
```

```python
df = pd.read_json('flipkart_fashion_products_dataset.json')
df.head()
```

```python
df['actual_price'].str.replace(',','')
```

```python
def clean_price(value):
    if not value:
        return np.nan
    elif isinstance(value, str):
        value = value.replace(',','')
        if value.isnumeric():    
            return float(value)
    else:   
      return value

df['actual_price'].apply(clean_price)
```

```python
df['actual_price'] = df['actual_price'].apply(clean_price)# actual implementation
df['selling_price'] = df['actual_price'].apply(clean_price)# actual implementation
```

```python
df.head()
```

```python
def clean_discount(value):
    if not value:
        return np.nan
    elif isinstance(value, str):
        value = value.replace('% off','')
        try:
            value = float(value)
            return value
        except:
            print(value)# get an idea of what value failed
            return np.nan
        else:
            return value
        
df['discount'].apply(clean_discount)       
```

```python
df['discount'] = df['discount'].apply(clean_discount)
```

```python
def clean_rating(value):
    if value:
        return float(value)
    return np.nan

#test
df['average_rating'].apply(clean_rating)
```

```python
df['average_rating'] = df['average_rating'].apply(clean_rating) #actual implementation
```
